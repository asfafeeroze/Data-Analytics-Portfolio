# What is the pattern of Churning?

```python
# Which gender churned more?

churned_customers = data.loc[data["Exited"] == 1] #rows you want, column to select and show. Not writing column name/range shows all columns.

churned_customers
```

```python
# What % churned? 

churned_percentage = data["Exited"].value_counts(normalize=True)

churned_percentage
```

```python


churned_customers["Gender"].value_counts().get("Female", 0)
```

```python
churned_customers["Gender"].value_counts().get("Male", 0)
```

```python
# Gender counts
gender_counts = churned_customers["Gender"].value_counts()

gender_counts.plot.pie(
    autopct='%1.1f%%',
    colors=['pink', 'lightblue'],
    startangle = 90
)

plt.title("Bank Churn Distribution by Gender")
plt.ylabel("")
plt.show()
```

```python
# Churn rate per gender

churned_customers.groupby('Gender')['Exited'].mean()
```

Although females represent a higher proportion of churned customers, the churn rate between genders is relatively similar, suggesting gender is not a strong predictor of churn.

```python
# Does age affect churn?

sns.boxplot(x="Exited", y="Age", data=data)

```

Based on the above boxplot, older customers churned more.

```python
# Plotting churning by country

geography_counts = churned_customers["Geography"]

print(churned_customers["Geography"].value_counts())

geography_counts.hist()

```

```python
# Does Geography Matter?

data.groupby('Geography')['Exited'].mean()
```

Customers in Germany have the highest churn rate.

```python
# Does balance affect Churn?

sns.boxplot(x="Exited", y="Balance", data=data)
```

People with higher balances are more likely to leave. 

```python
# Does number of products matter?

product_churn = data.groupby('NumOfProducts')['Exited'].mean()

product_churn.plot(kind="bar")

plt.title("Churn Rate by Number of Product")
plt.xlabel("Number of Products")
plt.ylabel("Churn Rate")
plt.show()
```

Based on average, people with 3 and more products are most likely to churn.

```python
sns.barplot(x=product_churn.index, y=product_churn.values)
```
